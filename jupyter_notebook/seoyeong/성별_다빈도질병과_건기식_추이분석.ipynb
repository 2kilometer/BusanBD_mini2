```python
year = [2019,2020,2021]

# for i in year :
#     file_path = "../seongho/file/질병분류/질병분류_여자_"+str(i)+".xlsx"
#     df=pd.read_excel(file_path,usecols="B:C")
#     df_tot = df.merge(df,on='상병명',how='left')
df_2019 = pd.read_excel("../seongho/file/질병분류/질병분류_여자_2019.xlsx",usecols="B:C")
df_2020 = pd.read_excel("../seongho/file/질병분류/질병분류_여자_2020.xlsx",usecols="B:C")
df_2021 = pd.read_excel("../seongho/file/질병분류/질병분류_여자_2021.xlsx",usecols="B:C")
```

```python
df_1920=df_2019.merge(df_2020,on='상병명', how='left')
df_wtot=df_1920.merge(df_2021,on='상병명', how='left')
```

```python
df_wtot=df_wtot.rename(columns={'진료실인원 (명)_x':'2019',
                            '진료실인원 (명)_y':'2020',
                            '진료실인원 (명)': '2021'})
```

```python
# 데이터 공백 제거
df_wtot['상병명'] = df_wtot['상병명'].str.replace(" ", "")
```

```python
for i in range(len(df_wtot)) :
    if str(df_wtot.loc[i,'상병명'])[0:1]=='K':
        df_wtot.loc[i,'K']='O'
```

```python
df_wk=df_wtot[df_wtot['K']=='O']
df_wk
```

```python
tot_sum=pd.DataFrame()

tot_sum['2019']=df_wk.groupby('K')['2019'].sum()
tot_sum['2020']=df_wk.groupby('K')['2020'].sum()
tot_sum['2021']=df_wk.groupby('K')['2021'].sum()
```
