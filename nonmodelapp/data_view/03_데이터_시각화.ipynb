```python
### 엔스콤 데이터를 이용한 시각화
import seaborn as sns

ans = sns.load_dataset("anscombe")
ans
```

```python
### dataset 데이터 중에 I 값만 추출하기
# 변수이름 : data1
data1 = ans[ans["dataset"] == "I"]
data1
```

```python
### 선그래프 그리기
# - data1의 x 컬럼은 x축,  y 컬럼은 y축으로 사용
import matplotlib.pyplot as plt

### 한글처리
plt.rc("font", family="Malgun Gothic")

### 마이너스 기호 사용 설정
plt.rcParams["axes.unicode_minus"] = False
```

```python
plt.plot(data1["x"], data1["y"])
```

```python
### 점으로 표현하기 : 산점도(분포 형태 확인 가능)
# - 소문자 : o
# - 산점도 : 데이터의 분포 및 흐름 확인 가능
#          : 우상향, 우하향, 선형, 비선형 등 확인 가능
plt.plot(data1["x"], data1["y"], "o")
```

```python
plt.plot(data1["x"], data1["y"], "x")
```

```python
### 4개의 각 그래프를 하나의 그래프로 표현하기 
```

```python
### 그래프 상위 객체 변수에 담기
fig = plt.figure()
```

```python
### 4개의 그래프가 들어갈 수 있는 공간 만들기
# - add_subplot(행, 열, 위치번호) : 작은 공간 만들기
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
fig
```

```python
### 각 그래프에 제목 넣기
ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")
fig
```

```python
### 그래프 layer 정렬하기
fig.tight_layout()
fig
```

```python
### 각각의 공간에 데이터를 넣어서 그래프 그리기
ax1.plot(data1["x"], data1["y"], "o", c="b")
ax2.plot(data2["x"], data2["y"], "o", c="r")
ax3.plot(data3["x"], data3["y"], "o", c="g")
ax4.plot(data4["x"], data4["y"], "o", c="y")

### 각 subplot에 x축 / y축 이름 넣기
ax1.set_xlabel("x-value")
ax1.set_ylabel("y-value")

ax2.set_xlabel("x-value")
ax2.set_ylabel("y-value")

ax3.set_xlabel("x-value")
ax3.set_ylabel("y-value")

ax4.set_xlabel("x-value")
ax4.set_ylabel("y-value")

### 전체 그래프 제목
fig.suptitle("Anscombe Data")

### 전체 x축/y축 이름넣기
fig.supxlabel("x-Value")
fig.supylabel("y-Value")

### 정렬
fig.tight_layout()

fig
```

```python
plt.title("시각화 종합")

x = [10, 20, 30, 40]
y = [1, 4, 9, 16]

# c : 색상
# lw : 선의 굵기
# ls : 선스타일
# marker : 지점 모양
# ms : 마커 사이즈
# mec : 마커 테두리 색상
# mew : 마커 테두리 두께
# mfc :마커 내부 채우기
plt.plot(x, y, 
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")

## x축의 범위
plt.xlim(0, 50)

plt.ylim(-10, 30)

plt.show()
```

```python
### tips 데이터 사용하기
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

### 데이터프레임(DataFrame)에서 제공하는 시각화

```python
### 데이터프레임에서 제공하는 시각화 plot() 함수 사용
# - 숫자값들만 처리됨

### 선그래프
tips.plot()
# - kind="line" : 디폴트
tips.plot(kind="line")

### 히스토그램 그래프
tips.plot(kind="hist")

### 박스플롯
tips.plot(kind="box")
```

### matplotlib에서 제공하는 시가화

```python
### 선그래프
plt.plot(tips.total_bill)
```

```python
plt.plot(tips["total_bill"])
```

```python
### 막대그래프 : 빈도 그래프
plt.bar(tips["sex"], tips["total_bill"])
```

```python
### 산점도 그래프
# - total_bill과 tip과의 관계에 따른 분포 확인하기
plt.scatter(tips["total_bill"], tips["tip"])
```

```python
### 히스토그램
# - total_bill 데이터별 빈도 시각화
plt.hist(tips["total_bill"])
```

```python
plt.hist(tips["day"])
```

### seaborn 시각화 (고급시각화에 속함)

```python
### 색상 변경하기
# 기본 팔레트 색상 : deep, muted, pastel, bright, dark, colorblind
sns.set_palette("pastel")

sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
### 막대 그래프
### - 오차막대(errorbar) : 비교 데이터간의 차이를 표시함
sns.barplot(x="sex", y="tip", data=tips)
```

```python
### - 오차막대를 안보이게 하려면 : errorbar=None 추가
sns.barplot(x="sex", y="tip", data=tips, errorbar=None)
```

```python
### 3개 데이터
sns.barplot(x="sex", y="tip", hue="day", data=tips)
```

```python
sns.boxplot(x="sex", y="tip", data=tips)
```

```python
sns.boxplot(data=tips, orient="h")
```

```python
### 커널 밀도
sns.kdeplot(tips["tip"])
```

```python
sns.distplot(tips["tip"])
```

```python
### 하나의 컬럼에 속한 범주 데이터를 count한 결과를 시각화함
sns.countplot(x="day", data=tips)
```

```python
### 두개 그래프 조합하여 그리기 : jointplot()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
### 다변량 시각화
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex")
```
