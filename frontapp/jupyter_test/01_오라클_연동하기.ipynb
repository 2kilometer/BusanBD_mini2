# DB 드라이버 연결

```python
dsn = cx_Oracle.makedsn('localhost', 1521, 'xe')
dsn
```

# DB 접속(계정/패스워드)

```python
conn = cx_Oracle.connect('busan', 'dbdb', dsn)
conn
```

# 커서 생성 (응답과 요청을 수행)

```python
### SQL 구문을 실어서 오라클에게 전달하고 응답결과를 받아오는 역할수행
cursor = conn.cursor()
cursor
```

# SQL 요청 및 응답결과 받아오기

```python
### SQL 구문 작성
sql = "Select * From member"

### SQL 구문을 커서(cursor)에 실어서 보내기(요청하기)
# - execute() : 요청함수
# - 결과는 cursor가 받아서 가지고 있음
cursor.execute(sql)
```

# 결과 출력

```python
### 받아온 모든 결과 출력하기
# - 응답결과 전체 출력 : fetchall() -> Select 결과가 1개행 이상인 경우
#   : 출력결과는 리스트[]에 튜플() 형태 (튜플 1개가 조회결과 중 행(row) 1개를 의미)
# - 응답결과 한건 출력 : fetchone() -> Select 결과가 1개행인 경우
#   : 출력결과는 튜플()형태 (행 1개)
rows = cursor.fetchall()
rows
```

```python
colnames = cursor.description
colnames
```

# DB 접속끊기(커서반환, 접속종료)

```python
### 접속 끊는 순서 : cursor 먼저 종료 > 다음에 conn 종료
cursor.close()
conn.close()
```

```python
### 컬럼명
#print(colnames)

### 받아온 데이터 확인하기->[(value, value), ()]
rows

### WEB에서 사용하는 형태는 [{key:value, key:value}, {}] 
#   리스트에 딕셔너리 형태로 변경해야함...

### 최종 변수명 : list_dict

# 1. 컬럼을 리스트로 변경하기
# 순수 for문으로만..
cols = []
for col in colnames:
    # print(col[0].lower())
    cols.append(col[0].lower())

list_dict = []
for row in rows :
    temp_dict = {}
    
    for i in range(len(cols)) :
        temp_dict[cols[i]] = row[i]
    # print(temp_dict)
    
    list_dict.append(temp_dict)

print(len(list_dict))    
list_dict
```
