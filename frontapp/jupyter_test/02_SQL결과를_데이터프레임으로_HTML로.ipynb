```python
"""
connect와 cursor는 매번 접속과 종료를 반복해야함
- 함수로 만들어서 사용..
- 자유롭게 함수이름 만든 후 사용...
- 생성한 함수를 이용해서 위에서 사용한 SQL구문을 사용하여 출력해보기
"""
```

```python
### SQL구문을 받아서 -> 데이터프레임을 리턴해주는 함수 생성
dsn = cx_Oracle.makedsn('localhost', 1521, 'xe')

def getDataFrame(sql) :
    conn = cx_Oracle.connect('busan', 'dbdb', dsn)
    cursor = conn.cursor()
    
    cursor.execute(sql)
    rows = cursor.fetchall()
    
    colnames = cursor.description
    cols_list = [colnames[i][0] for i in range(len(colnames))]
    
    cursor.close()
    conn.close()
    
    return pd.DataFrame(rows, columns = cols_list)
```

```python
### 함수호출하기
sql = """
    Select mem_id, mem_name, 
            Substr(mem_add1, 1, 2) as area,
            Substr(cart_no, 1, 8) as cart_date,
            cart_qty
    From member, cart
    Where mem_id = cart_member
    Order By mem_id Asc
"""
df4 = getDataFrame(sql)
df4
```

```python
"""
sql = 주문을 한번도 하지 않은 회원의 아이디, 이름 조회하기
- 데이터프레임 변수명 : df5
"""
sql = """
    Select mem_id, mem_name
    From member
    Where Not Exists(Select cart_member From cart 
                                    Where cart_member = mem_id)
"""
df5 = getDataFrame(sql)
df5
```

```python
"""
sql = 회원 정보 조회하기, 단, 아이디 a001 회원에 대해서
"""
sql = """
    Select *
    From member
    Where mem_id = 'a001'
"""
df6 = getDataFrame(sql)
df6
```

```python
def initConn():
    conn = cx_Oracle.connect('busan', 'dbdb', dsn)
    cursor = conn.cursor()
    return conn, cursor

def getDataFrame2(conn, cursor):
    rows = cursor.fetchall()

    colnames = cursor.description
    cols_list = [colnames[i][0] for i in range(len(colnames))]

    cursor.close()
    conn.close()

    return pd.DataFrame(rows, columns = cols_list)
```

```python
### 함수 분리..
# 분리된 함수로 위의 sql 구문 실행해서 데이터프레임으로 출력까지 해주세요..
```

```python
conn, cursor = initConn()

sql = """
    Select *
    From member
    Where mem_id = :mem_id
      And mem_mileage >= :mem_mileage
"""
cursor.execute(sql, mem_id='b001', mem_mileage=1000)

df7 = getDataFrame2(conn, cursor)
df7
```

```python
id = "a001"
mileage = 1000

sql = """
    Select *
    From member
    Where mem_id = '{}'
      And mem_mileage >= {}
""".format(id, mileage)

df6 = getDataFrame(sql)
df6
```

```python
"""
회원 거주지역별 주문수량의합을 조회하기
- 정렬은 거주지역 오름차순
- 상품분류명에 "전자제품"이 포함되고
- 거래처명에 "대우전자"가 포함된 경우
- 함수는 편하신 것으로 사용
- 데이터프레임 변수명 : df8
"""

sql = """
    Select substr(mem_add1, 1, 2) as area, 
           Sum(cart_qty) as sum_qty
    From member, cart, prod, lprod, buyer
    Where mem_id = cart_member
       And prod_id = cart_prod
       And lprod_gu = prod_lgu
       And buyer_id = prod_buyer
       And lprod_nm Like :lp_name
       And buyer_name Like :b_name
    Group By substr(mem_add1, 1, 2)
    Order By area Asc
"""
conn, cursor = initConn()
cursor.execute(sql, lp_name='%전자제품%', b_name='%대우전자%')
df8 = getDataFrame2(conn, cursor)
df8
```

```python
sql = """
    Select substr(mem_add1, 1, 2) as area, 
           Sum(cart_qty) as sum_qty
    From member, cart, prod, lprod, buyer
    Where mem_id = cart_member
       And prod_id = cart_prod
       And lprod_gu = prod_lgu
       And buyer_id = prod_buyer
       And lprod_nm Like '%{}%'
       And buyer_name Like '%{}%'
    Group By substr(mem_add1, 1, 2)
    Order By area Asc
""".format("전자제품", "대우전자")
df8 = getDataFrame(sql)
df8
```

```python
### 여러개의 매개변수를 한번에 받아오는 방법
# - **params : 여러개의 값을 받아서 딕셔너리로 만들어냄

def getDataFrame3(sql, **params) :
    print(params)
    conn = cx_Oracle.connect('busan', 'dbdb', dsn)
    cursor = conn.cursor()
    
    cursor.execute(sql, params)
    rows = cursor.fetchall()
    
    colnames = cursor.description
    cols_list = [colnames[i][0] for i in range(len(colnames))]
    
    cursor.close()
    conn.close()
    
    return pd.DataFrame(rows, columns = cols_list)
```

```python
sql = """
    Select substr(mem_add1, 1, 2) as area, 
           Sum(cart_qty) as sum_qty
    From member, cart, prod, lprod, buyer
    Where mem_id = cart_member
       And prod_id = cart_prod
       And lprod_gu = prod_lgu
       And buyer_id = prod_buyer
       And lprod_nm Like :lp_name
       And buyer_name Like :b_name
    Group By substr(mem_add1, 1, 2)
    Order By area Asc
"""

df8 = getDataFrame3(sql, lp_name="%전자제품%", b_name="%대우전자%")
df8
```

```python
"""
회원성별, 나이대, 상품명, 주문수량의합 조회하기
- 성별은 "남자" 또는 "여자"로 조회
- 나이대 : 10대미만, 10대, 20대, ...70대, 80대이상, 80대이상..형태로 조회
"""

sql = """
    Select Decode(Mod(Substr(mem_regno2, 1, 1), 2), 0, '여자', '남자') as fm,
           (Case 
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 9 Then '10대미만'
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 19 Then '10대'
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 29 Then '20대'
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 39 Then '30대'
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 49 Then '40대'
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 59 Then '50대'
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 69 Then '60대'
                When Extract(year from sysdate) - Extract(year from mem_bir) <= 79 Then '70대'
                Else '80대이상'
           End) as age,
           prod_name, Sum(nvl(cart_qty, 0)) as sum_qty
    From member, cart, prod
    Where mem_id = cart_member
       And prod_id = cart_prod
    Group By Decode(Mod(Substr(mem_regno2, 1, 1), 2), 0, '여자', '남자'),
               (Case 
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 9 Then '10대미만'
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 19 Then '10대'
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 29 Then '20대'
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 39 Then '30대'
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 49 Then '40대'
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 59 Then '50대'
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 69 Then '60대'
                    When Extract(year from sysdate) - Extract(year from mem_bir) <= 79 Then '70대'
                    Else '80대이상'
               End),
               prod_name
"""
df9 = getDataFrame(sql)
df9
```
