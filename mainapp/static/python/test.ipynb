```python
import pandas as pd
import cx_Oracle
from sqlalchemy import create_engine

# 엑셀 파일을 데이터프레임으로 읽기
df = pd.read_csv('../../../disease.csv' , encoding='cp949')
```

```python
df
```

```python
age_group_mapping = {
    '1 - 9세': '10세 미만',
    '10 - 19세': '10대',
    '20 - 29세': '20대',
    '30 - 39세': '30대',
    '40 - 49세': '40대',
    '50 - 59세': '50대',
    '60 - 64세': '60대',
    '65 - 69세': '60대',
    '75세 이상': '70대 이상',
}

# Replace the age groups in the relevant columns
df['Most_Common_Age_Group'] = df['Most_Common_Age_Group'].map(age_group_mapping)
df['2nd_Most_Common_Age_Group'] = df['2nd_Most_Common_Age_Group'].map(age_group_mapping)
df['3rd_Most_Common_Age_Group'] = df['3rd_Most_Common_Age_Group'].map(age_group_mapping)

df.head()
```

```python
df.to_csv('../../../disease.csv', index=False, encoding='cp949')
```
